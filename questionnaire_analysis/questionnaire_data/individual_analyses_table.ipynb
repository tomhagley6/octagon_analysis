Socials

```python
# trial logic
scores_per_unit_time = [(0, 0) for _ in range(len(social_lists))]
delta_times_list = []
for trial_list_index in range(len(social_lists)):
    # define trial
    trial_list = social_lists[trial_list_index]
    delta_times = []
    for trial_index in range(len(trial_list)):
        trial = trial_list[trial_index]
        # cut trials to only return timepoints between slice onset and server-selected trigger activation
        slice_onset = trial[trial['eventDescription'] == globals.SLICE_ONSET]
        slice_onset_index = slice_onset.index[0] - trial.index[0]
        selected_trigger_activation = trial[trial['eventDescription'] == globals.SELECTED_TRIGGER_ACTIVATION]
        selected_trigger_activation_index = selected_trigger_activation.index[0] - trial.index[0]

        # extract application time between slice onset and server-selected trigger activation
        time_app_onset = trial['timeApplication'].iloc[slice_onset_index]
        time_app_trigger = trial['timeApplication'].iloc[selected_trigger_activation_index]        
        time = time_app_trigger - time_app_onset
        delta_time = time.total_seconds()
        delta_times.append(delta_time)
        
    total_time = sum(delta_times)
    delta_times_list.append(total_time)

    # access final trial event log event for the final player scores
    final_trial = trial_list[-1]
    final_trial_trial_end = final_trial[final_trial['eventDescription'] == 'trial end']
    
    # flexibly index player scores
    player0_score = final_trial_trial_end[globals.PLAYER_SCORE_DICT[0]['score']].item()
    player1_score = final_trial_trial_end[globals.PLAYER_SCORE_DICT[1]['score']].item()

    # score per unit time as final score divided by total time spent playing (active trials)
    score_per_time_unit_p0 = player0_score/total_time
    score_per_time_unit_p1 = player1_score/total_time
    print(player0_score, player1_score)
    print(score_per_time_unit_p0, score_per_time_unit_p1)
    scores_per_unit_time[trial_list_index] = (score_per_time_unit_p0, score_per_time_unit_p1)
```

```python
scores_per_unit_time_dict = {}

for i, tuple in enumerate(pseudo_tuples_list):
    player0 = tuple[0]
    player1 = tuple[1]

    if player0 not in scores_per_unit_time_dict:
        scores_per_unit_time_dict[player0] = []
    if player1 not in scores_per_unit_time_dict:
        scores_per_unit_time_dict[player1] = []
    
    scores_per_unit_time_dict[player0] = scores_per_unit_time[i][0]
    scores_per_unit_time_dict[player1] = scores_per_unit_time[i][1]
```
